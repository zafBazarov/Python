# Load sample data

**Load CSV File**

I'm going to show you how to load CSV and Excel files using the Pandas library. So the first thing you have to do is you have to find out where your files are located. I have two files, car_financing.csv and car_financing.xlsx. In this cell, we have a path to the file car_financing.csv. And what this here is is a relative path. So relative to the notebook load_data.ipynb I have the Data folder and inside the Data folder is the CSV file that we want to load. From here, we're going to use the pd.read_csv method to load that file into a Pandas DateFrame. And I'm going to press Shift + Enter. 

```python
# Import libraries
import pandas as pd
```

```python
# Load car loan data from a scv file

df = pd.read_csv('car_financing.csv')
df
```

**Load Excel File**

You can also load files from Excel. We have a folder called Data, and inside that folder is a file called car_financing.xlsx. And what we're going to do now is we're going to read the Excel file using the method pd_read_excel into a Pandas DateFrame. And like before, we've loaded the file into a Pandas DateFrame. 

```python
filename = 'car_financing.xlsx'
df = pd.read_excel(filename)
df
```
